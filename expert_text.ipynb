```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load a transformer-based text model from TensorFlow Hub (Updated Variation)
model_url = "https://tfhub.dev/google/universal-sentence-encoder-large/5"
embed = hub.KerasLayer(model_url, trainable=False)
```

```python
# Example text data (Variation: Different domain-specific sentences)
texts = [
    "Blockchain technology is transforming the financial sector.",
    "Cryptographic security ensures data integrity in digital transactions.",
    "Neural networks enable state-of-the-art performance in NLP.",
    "Self-supervised learning is a promising research direction."
]
```

```python
# Convert text to embeddings
embeddings = embed(texts)
```

```python
# Compute cosine similarity between sentences
similarity_matrix = cosine_similarity(embeddings.numpy())
```

```python
# Plot the similarity matrix
plt.figure(figsize=(8, 6))
sns.heatmap(similarity_matrix, annot=True, fmt=".2f", xticklabels=texts, yticklabels=texts, cmap="coolwarm")
plt.title("Cosine Similarity Between Text Embeddings")
plt.show()
```
